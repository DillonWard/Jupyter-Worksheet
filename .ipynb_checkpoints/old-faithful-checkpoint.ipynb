# Reading in from CSV

```python
import numpy as np
import csv

with open('./data/faithful.csv') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    eruptions = []
    waiting = []
    for row in readCSV:
        eruptions = row[0]
        waiting = row[1]

        eruptions.append(eruptions)
        waiting.append(waiting)
```
